```python
def world_map(locations,counts,title):
    data = [ dict(
            type = 'choropleth',
            locations = locations,
            z = counts,
            colorscale = 'Blues',
            locationmode = 'country names',
            autocolorscale = False,
            reversescale = True,
            marker = dict(
                line = dict(color = '#F7F7F7', width = 1.5)),
                colorbar = dict(autotick = True, legth = 3, len=0.75, title = 'respodents',
                               max = 1000, min = 0)
                )
           ]
    layout = dict(
        title = title,
        titlefont={'size': 28, 'family': 'san serif'},
        width=750, 
        height=475,
        paper_bgcolor='#F7F7F7',
        geo = dict(
            showframe = True,
            showcoastlines = True,
            fitbounds="locations",
            )
    )
    
    fig = dict(data=data, layout=layout)
    iplot(fig, validate=False, filename='world-map')
    
z = df21_Ea['Q3'].value_counts()
 
## 메서드 호출
world_map(locations=z.index, counts=z.values, title= '<b> EastAsia Countries (2021 survey) <b>')
```

```python
#Gender 뽑아오기
gender21= df21_Ea.loc[:, ['Q3', 'Q2', 'year']].rename(columns={'Q3':'Country', 'Q2':'Gender'})
gender20= df20_Ea.loc[:, ['Q3', 'Q2', 'year']].rename(columns={'Q3':'Country', 'Q2':'Gender'})
gender19= df19_Ea.loc[:, ['Q3', 'Q2', 'year']].rename(columns={'Q3':'Country', 'Q2':'Gender'})
gender18= df18_Ea.loc[:, ['Q3', 'Q1', 'year']].rename(columns={'Q3':'Country', 'Q1':'Gender'})
gender17= df17_Ea.loc[:, ['Country', 'GenderSelect', 'year']].rename(columns={'index':'type', 'GenderSelect':'Gender'})

#data 정제(한꺼번에 이름바꾸기)
Gender5y= pd.concat([gender17, gender18, gender19, gender20, gender21])
Gender5y= (Gender5y.replace(['Prefer not to say', 'Prefer to self-describe', 'Nonbinary', 'A different identity'], 'Others')
           .replace(['Man', 'Woman'], ['Male', 'Female'])
           .groupby(['year', 'Gender'])
           .size()
           .reset_index()
           .rename(columns = {0:"Count"}))

#percent data 넣기
gen17_percent = Gender5y[Gender5y['year'] == "2017"].reset_index(drop = True)
gen17_percent['percentage'] = gen17_percent["Count"] / gen17_percent["Count"].sum()
gen17_percent['%'] = np.round(gen17_percent['percentage'] * 100, 1)

gen18_percent = Gender5y[Gender5y['year'] == "2018"].reset_index(drop = True)
gen18_percent['percentage'] = gen18_percent["Count"] / gen18_percent["Count"].sum()
gen18_percent['%'] = np.round(gen18_percent['percentage'] * 100, 1)

gen19_percent = Gender5y[Gender5y['year'] == "2019"].reset_index(drop = True)
gen19_percent['percentage'] = gen19_percent["Count"] / gen19_percent["Count"].sum()
gen19_percent['%'] = np.round(gen19_percent['percentage'] * 100, 1)

gen20_percent = Gender5y[Gender5y['year'] == "2020"].reset_index(drop = True)
gen20_percent['percentage'] = gen20_percent["Count"] / gen20_percent["Count"].sum()
gen20_percent['%'] = np.round(gen20_percent['percentage'] * 100, 1)

gen21_percent = Gender5y[Gender5y['year'] == "2021"].reset_index(drop = True)
gen21_percent['percentage'] = gen21_percent["Count"] / gen21_percent["Count"].sum()
gen21_percent['%'] = np.round(gen21_percent['percentage'] * 100, 1)
gen21_percent

#data 완성
Gen5y_percent = pd.concat([gen17_percent, gen18_percent, gen19_percent, gen20_percent, gen21_percent], ignore_index = True)
Gen5y_percent= pd.pivot(Gen5y_percent, index = "year", columns = "Gender", values = "%").reset_index()
Gen5y_percent
```
